Hago el análisis de los 2 primeros PC

```python
from sklearn.decomposition import PCA
 
pca = PCA(2)  #me quedo con las 2 componentes principales.
PC_flowers = pca.fit_transform(df_array)
 
```

```python
#creo el array con las clases de cada flor
target_flo=np.array(list(labels['label']))
```

```python
# Grafico los puntos en los 2 componentes pricipales y les pongo color de acuerdo a la clase

plt.style.use('seaborn-whitegrid')
fig_1=plt.figure(figsize = (10,7))
c_map = plt.cm.get_cmap('jet', 10)
plt.scatter(PC_flowers[:, 0], PC_flowers[:, 1], s = 15,
            cmap = c_map , c = target_flo)
#plt.colorbar()
plt.xlabel('PC-1') , plt.ylabel('PC-2')
#plt.title('2 Componentes principales')
plt.show()
```

Hago el análisis de los primeros 3 PC

```python
pca_3 = PCA(3) # me quedo con 2 componentes principales.
PC_flowers_3 = pca_3.fit_transform(df_array)
 
import matplotlib.pyplot as plt

fig_2=plt.figure(figsize=(10, 7))
ax = fig_2.add_subplot(projection='3d')

c_map = plt.cm.get_cmap('jet', 10)

sequence_containing_x_vals = list(PC_flowers_3[:,0])
sequence_containing_y_vals = list(PC_flowers_3[:,1])
sequence_containing_z_vals = list(PC_flowers_3[:,2])

ax.scatter(sequence_containing_x_vals, sequence_containing_y_vals, sequence_containing_z_vals, cmap= c_map, c = target_flo)
ax.set_xlabel('PC-1') , ax.set_ylabel('PC-2'), ax.set_zlabel('PC-3')
ax.view_init(-130, 60)
#plt.title('3 Componentes principales')
plt.show()
```

Cuanta varianza explican estos PC?

```python
pca = PCA(100)
#
# Determine transformed features
#
df_array_pca = pca.fit_transform(df_array)
#
# Determine explained variance using explained_variance_ration_ attribute
#
exp_var_pca = pca.explained_variance_ratio_
#
# Cumulative sum of eigenvalues; This will be used to create step plot
# for visualizing the variance explained by each principal component.
#
cum_sum_eigenvalues = np.cumsum(exp_var_pca)
#
# Create the visualization plot
#
plt.bar(range(0,len(exp_var_pca)), exp_var_pca, alpha=0.5, align='center', label='Individual explained variance')
plt.step(range(0,len(cum_sum_eigenvalues)), cum_sum_eigenvalues, where='mid',label='Cumulative explained variance')
plt.ylabel('Explained variance ratio')
plt.xlabel('Principal component index')
plt.legend(loc='best')
plt.tight_layout()
plt.show()
```
